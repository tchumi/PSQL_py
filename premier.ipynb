```python
import pandas as pd
```

```python
df = pd.read_csv('declarations.csv',on_bad_lines='skip',sep=';', nrows=10000)  # Charge 10 000 premières lignes
print(df.head())
```

```python
df.shape
```

```python
print(df.columns.sort_values())
```

```python
df_col_triees = df[df.columns.sort_values()]
df_col_triees.to_csv('declarations_col_triees.csv', sep=';', index=False)
```

```python
# suppression des doublons
df = df.drop_duplicates()
```

```python
for col in df.columns.sort_values():
    nb_unique = len(df[col].unique())
    if nb_unique < 20:
        print(col, len(df[col].unique()), df[col].unique())
```

```python
# export du fichier csv sans doublons
df.to_csv('declarations_sans_doublons.csv', sep=';', index=False)
```
