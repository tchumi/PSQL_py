```python
expected_columns = 55

# Ouvrir le fichier CSV et le fichier d'erreurs avec l'encodage approprié
with open('declarations_col_triees.csv', 'r', encoding='utf-8') as csv_file, open('erreur.txt', 'w', encoding='utf-8') as error_file:
    for i, line in enumerate(csv_file):
        columns = line.split(';')  # Diviser la ligne en colonnes
        if len(columns) != expected_columns:
            # Écrire un message dans le fichier d'erreurs
            error_file.write(f"Ligne {i + 1} : {len(columns)} colonnes au lieu de {expected_columns}\n")
```
